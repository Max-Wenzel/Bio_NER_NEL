```python
import sys

from collections import defaultdict as dd

sys.path.append("../NER")

from crf_funcs import *
```

```python
test = file_opener("../NER/test")

dev = file_opener("../NER/dev")

train = file_opener("../NER/train")
```

```python
all_entries = train# + test + dev
```

```python
len(all_entries)
```

```python
tag_phrases = dd(set)
from nltk.stem import PorterStemmer
ps = PorterStemmer()

for sentence in all_entries:
    in_phrase = False
    phrase = [[], '']
    for word in sentence:
        if word[2][0] == "B":
            if in_phrase:
                    tag_phrases[phrase[1]].add(tuple(phrase[0]))
                    phrase = [[], '']
            phrase[0].append(word[0].lower())
            phrase[1] = word[2][2:]
            in_phrase = True
        elif word[2][0] == "I":
            phrase[0].append(word[0].lower())
        else:
            if in_phrase:
                tag_phrases[phrase[1]].add(tuple(phrase[0]))
                phrase = [[], '']
                in_phrase = 0
```
